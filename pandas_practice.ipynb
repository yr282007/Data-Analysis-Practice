```python
from datetime import date, timedelta

startingDate = date(2019, 7, 1)
endingDate= date(2020,6,30)
diff = endingDate - startingDate
```

```python
dates = []
for k in range(diff.days+1):
    dates.append(startingDate + timedelta(days = k))

dates
```
